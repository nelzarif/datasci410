<h2>Relationship 1: Heating and Cooling Load Distribution</h2>
    
We will take a look at the distributions of heating and cooling loads. The histograms show us patterns have a similar shape

```python
energy_data.loc[:, "Heating Load"].plot.hist(bins=10)
```

```python
energy_data.loc[:, "Cooling Load"].plot.hist(bins=10)
```

By observing histograms of heating and cooling loads, we can see that they follow the same pattern. From this We can conclude that buildings that require alot of energy for heating also require a lot of energy for cooling. Maintaing temperature is a challeneg on both ends of the spectrum.

<h2>Relationship 2: Orientation & Overall Height Effect On Heating/Cooling Load</h2>

I will investigate orientation's effect on the heating cooling factor. Then I will investigate Overall Height's effect on heating/cooling load

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6,6))
ax = fig.gca()
energy_data.loc[:, ['Orientation', 'Heating Load']].boxplot(by= 'Orientation', ax=ax)
ax.set_title('Heating load by Orientation')
ax.set_ylabel('Heating Load')
```

This boxplot shows us that orientation has no noticible effect on the heating load.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.gca()
energy_data.loc[:, ['Overall Height', 'Heating Load']].boxplot(by= 'Overall Height', ax=ax)
ax.set_title('Heating load by Overall Height')
ax.set_ylabel('Heating Load')
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.gca()
energy_data.loc[:, ['Overall Height', 'Cooling Load']].boxplot(by= 'Overall Height', ax=ax)
ax.set_title('Cooling load by Overall Height')
ax.set_ylabel('Cooling Load')
```

The box plots show us that that higher structures have a bigger heating/cooling loads. There are more outlier for cooling loads than heating loads.

As we have speculated earlier, we can also see similar trends for heating/ cooling loads. This further proves our findings in relationship #1.

<h2>Relationship 3: Glazing Area</h2>


```python
import seaborn as sns
fig = plt.figure(figsize=(8,8))
ax = fig.gca()
sns.set_style("whitegrid")
sns.violinplot(x='Glazing Area', y= 'Heating Load', data = energy_data, ax=ax)
#sns.violinplot(x='Glazing Area' , y= 'Cooling Load', data = energy_data, ax=ax)
```

As the Glazing Area increases, the heating load increases

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca()
sns.set_style("whitegrid")
sns.violinplot(x='Glazing Area', hue='Overall Height', y= 'Heating Load', data = energy_data, ax=ax)
```

We can see the last findings reinforced in this plot. As the glazing area increases, the heating load increases. For every glazing area, we can see buildings with the larger overall height have higher heating loads
