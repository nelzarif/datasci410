# Set Up 

```python
# Cleaned up Dataset location
fileName = "https://library.startlearninglabs.uw.edu/DATASCI410/Datasets/Automobile%20price%20data%20_Raw_.csv"
```

```python
# Import libraries
import pandas as pd
import numpy as np
import scipy.stats as ss
from matplotlib import pyplot as plt
import statsmodels.stats.power as ssp
%matplotlib inline
```

```python
def read_auto_data(fileName = "https://library.startlearninglabs.uw.edu/DATASCI410/Datasets/Automobile%20price%20data%20_Raw_.csv"):
    'Function to load the auto price data set from a .csv file' 

    ## Read the .csv file with the pandas read_csv method
    auto_price = pd.read_csv(fileName)
    
    ## Remove rows with missing values, accounting for mising values coded as '?'
    cols = ['price', 'bore', 'stroke', 
          'horsepower', 'peak-rpm']
    for column in cols:
        auto_price.loc[auto_price[column] == '?', column] = np.nan
    auto_price.dropna(axis = 0, inplace = True)

    ## Convert some columns to numeric values
    for column in cols:
        auto_price[column] = pd.to_numeric(auto_price[column])
        
    return auto_price
auto_price = read_auto_data()
```

# Part 1: Testing Normality

## Normality of Price
The price distribution of the auto data set can be ploted to demonstarate normality. A simple histogram can show that the distribution of price is not normal.

## Kolmogorov-Smirnov test 

The KS test shows us that both, price and log price, are not normal. The KS statistic is very high, and the p-value is very low. This means we can reject the null hypothesis that these distributions are the same.

```python
def ks_test(dat, dist = 'norm'):
    from scipy.stats import kstest 
    ks_statisic, pvalue = kstest(dat, dist)
    print('KS-statistic = ' + str(ks_statisic))
    print('P-value = ' + str(pvalue))

print('Price KS Test')
ks_test(auto_price.loc[:,'price'], 'norm')

print('\nLog Price KS Test')
ks_test(auto_price.loc[:,'log_price'], 'norm')
```

# Testing Significance of Price

## Stratifying  Data
We want a large sample to get a good representation of the population, but there are only 195 rows. I will stratify the data, and make the size of the sample equal to the size of the smallest strata. 

```python
auto_price.shape
```

```python
def stratify_by(df, group):
    #getting the size of the type with the lowest count
    size = df[group].value_counts().min()
    print('The sample size of each strata in ' + str(group) + ' is ' + str(size))
    
    groups = df.groupby([group])
    return groups.apply(lambda x: x.sample(n=size))
```

## Testing Significance

```python
def t_test(a, b, alpha, alternative='two-sided'):
    import statsmodels.stats.weightstats as ws
    
    diff = a.mean() - b.mean()

    res = ss.ttest_ind(a, b, equal_var=False)
      
    means = ws.CompareMeans(ws.DescrStatsW(a), ws.DescrStatsW(b))
    confint = means.tconfint_diff(alpha=alpha, alternative=alternative, usevar='unequal') 
    degfree = means.dof_satt()

    index = ['DegFreedom', 'Difference', 'Statistic', 'PValue', 'Low95CI', 'High95CI']
    return pd.Series([degfree, diff, res[0], res[1], confint[0], confint[1]], index = index)   
   
```

```python
%matplotlib inline
def hist_plot(vec, bins):
    import matplotlib.pyplot as plt
    from statistics import mean
    plt.hist(vec, bins = bins)
    plt.axvline(mean(vec), color = 'red')
    plt.ylabel('Frequency')
    plt.xlabel('Value')

def hist_mean_ci(vec, t_test, bins):
    import matplotlib.pyplot as plt
    from statistics import mean
    upper = mean(vec) + t_test[5] - t_test[1]
    lower = mean(vec) + t_test[4] - t_test[1]
    plt.hist(vec, bins = bins)
    plt.axvline(mean(vec), color = 'red')
    plt.axvline(upper, color = 'red', linestyle='--')
    plt.axvline(lower, color = 'red', linestyle='--')
    plt.ylabel('Frequency')
    plt.xlabel('Value')

def plot_t(a, b, t_test, cols=['pop_A', 'pop_B'], nbins = 20):
    import matplotlib.pyplot as plt
    minx = min([min(a), min(b)])
    maxx = max([max(a), max(b)])
    stepx = (maxx - minx)/(nbins + 1)
    bins = [minx + i * stepx for i in range(nbins + 1)]
    plt.subplot(2, 1, 1) # The first plot
    hist_mean_ci(a, t_test, bins)
    plt.title('Histogram of ' + cols[0] + ' and ' + cols[1])
    plt.subplot(2, 1, 2) # The next plot
    hist_plot(b, bins)
```

```python
import statsmodels.stats.power as smsp

def plot_power(x, y, xlabel, title):
    import matplotlib.pyplot as plt
    plt.plot(x, y, color = 'red', linewidth = 2)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Power')
```

### Fuel Type 
This data is stratified by fuel type, gas or diesel.

```python
fuel_strata = stratify_by(auto_price, 'fuel-type')
fuel_strata.reset_index(inplace = True, drop=True)
```

```python
# I will drop all columns except for fuel-type and price
fuel_strata = fuel_strata[['fuel-type', 'price']]

diesel = fuel_strata.loc[fuel_strata.loc[:,'fuel-type'] == 'diesel','price']
gas = fuel_strata.loc[fuel_strata.loc[:,'fuel-type'] == 'gas','price']
```

Looking at the t-test below, we cannot reject the null hypothesis. The t-statistic is small, the p-value is greater than 0.05, and the confidence interval straddles 0. 

```python
test = t_test(diesel, gas, 0.05)
test
```

The plot below confims the numbers above. The mean of the second sample is within 95% confidence interval. We fail to reject the null hypothesis, and the differnce of means can arrise from randomness. We can't say with confidence that there is a significance in mean as a result of gas type.

```python
plot_t(diesel, gas, test, cols = ['diesel', 'gas'])
```

Given a difference of means of 3000, we can see that the power of the test is 1. A difference as small as $100 has a power of 1. 

```python
diffs = np.arange(start = 0.0, stop = 3000, step = 100)
powers = [smsp.tt_ind_solve_power(effect_size = x, nobs1 = 20, 
           alpha = 0.05, power = None, ratio = 1.0, alternative = 'two-sided') 
          for x in diffs]


plot_power(diffs, powers, xlabel = 'Difference of means', title = 'Power vs. difference of means')    
```

The effect size makes this test powerful, but if the difference of means was very small like 1 dollar difference, the sample size would not be sufficient. Looking at the plot below, we can see that at 20 observations, the power of the test is at 0.85 

```python
nobs = np.arange(start = 1.0, stop = 22, step = 1.0) 
powers = [smsp.tt_ind_solve_power(effect_size = 1, nobs1 = x,
           alpha = 0.05, power = None, ratio = 1.0, alternative = 'two-sided') 
          for x in nobs]
plot_power(nobs, powers, xlabel = 'Number of observations', title = 'Number of observations') 
```

## ANOVA

The f-statistic is large and the p-value is very small. This means that we can reject the null hypothesis that the difference in means can arise from randomness. 

```python
f_statistic, p_value = ss.f_oneway(convertible, hardtop, hatchback, sedan, wagon)
print('F statistic = ' + str(f_statistic))
print('P-value = ' + str(p_value))
```

Our effect size, which is the difference of means, is in the magnitude of hundreds. Even when our smallest observation is 6 samples big, this test has a power of 1 due to the scale of the difference of cost.

```python
diffs = np.arange(start = 0.0, stop = 300, step = 1) 
powers = smsp.FTestAnovaPower().solve_power(effect_size = diffs, nobs=6, alpha=0.05)
plot_power(diffs, powers, xlabel = 'Difference', title = 'Power vs. difference') 
```

## Tukey HS Test

The Tukey HS Test allows us to compare te different body styles by pairs, which is more accurate. The ANOVA test merely gave us a statistic that not all the styles have overlapping means.

The table below shows us 4 instances where we cannot reject the null hypothesis. In these rows, we can see the confidence interval straddling 0. 

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd = pairwise_tukeyhsd(auto_price.loc[:, 'price'], auto_price.loc[:, 'body-style'])
print(hsd)
```

Visually it is very easy to see there are 4 instances where the intervals overlap. We cannot reject the null hypothesis, that the difference of means arises from randomness, for pairs where the confidence intervals overlap.
